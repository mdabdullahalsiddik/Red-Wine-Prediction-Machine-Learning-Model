# Blearnce Data

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
#creating RandomOverSampler object
ros = RandomOverSampler()
```

```python
x_n,y_n = ros.fit_resample(x,y)
```

```python
y_n.value_counts()
```

# Training & Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
# 70% training data and 30% testing data
xtrain,xtest,ytrain,ytest = train_test_split(x_n,y_n,test_size=0.30,random_state=42)
```

```python
#training dataset 1st 5 row show
xtrain.head()
```

```python
#training dataset 1st 5 row show
ytrain.head()
```

```python
#training dataset size
xtrain.shape
```

```python
#training dataset size
ytrain.shape
```

```python
#testing dataset size
xtest.shape
```

```python
#testing dataset size
ytest.shape
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
#creating LogisticRegression object
logr = LogisticRegression()
```

```python
logr.fit(xtrain,ytrain)
```

```python
logr.predict(xtest)
```

# Accuracy Score & Classification Report & Confusion Matrix & Roc Curve

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,plot_roc_curve
```

```python
logr.score(xtrain,ytrain)
```

```python
logr.score(xtest,ytest)
```

```python
accuracy_score(ytest,logr.predict(xtest))
```

```python
classification_report(ytest,logr.predict(xtest))
```

```python
print(classification_report(ytest,logr.predict(xtest)))
```

```python
confusion_matrix(ytest,logr.predict(xtest))
```

```python
sns.heatmap(confusion_matrix(ytest,logr.predict(xtest)),annot=True)
```

```python
plt.figure(figsize=(5,5))
plot_roc_curve(logr,xtest,ytest)
plt.plot([0,1],[0,1])
plt.show()
```

# Download Machine Learning Model

```python
import pickle as pk
```

```python
pk.dump(dtc,open("Red Wine Prediction Machine Learning Model","wb"))
```

```python
model = pk.load(open("Red Wine Prediction Machine Learning Model","rb"))
```

```python
fixed_acidity = float(input("Enter Red Wine Fixed Acidity Valu = "))
volatile_acidity = float(input("Enter Red Wine Volatile Acidity Valu = "))
citric_acid = float(input("Enter Red Wine Citric Acid Valu = "))
residual_sugar = float(input("Enter Red Wine Residual Sugar Valu = "))
chlorides = float(input("Enter Red Wine Chlorides Valu = "))
free_sulfur_dioxide = float(input("Enter Red Wine Free Sulfur Dioxide Valu = "))
total_sulfur_dioxide = float(input("Enter Red Wine Total Sulfur Dioxide Valu = "))
density = float(input("Enter Red Wine Density Valu = "))
pH = float(input("Enter Red Wine PH Valu = "))
sulphates = float(input("Enter Red Wine Sulphates Valu = "))
alcohol = float(input("Enter Red Wine Alcohol Valu = "))

result = model.predict([[fixed_acidity,volatile_acidity,citric_acid,residual_sugar,chlorides,free_sulfur_dioxide,total_sulfur_dioxide,density,pH,sulphates,alcohol]])

if result == 1:
    print("Good Wine")
else:
    print("Not Good Wine")
```
